```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
x_train[0]
```

```python
plt.figure(figsize=(28,28))
plt.imshow(x_train[0],cmap='gray')
```

```python
x_train.shape
```

```python
x_train_final = x_train.reshape(len(x_train),784)
x_train_final = x_train_final/255
x_train_final.shape
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(10, activation='sigmoid', input_shape=(784,)))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train_final, y_train, epochs = 20)
```

```python
model.save_weights('./my_model.weights.h5')
```

```python
model.load_weights('./my_model.weights.h5')
```

```python
x_test_final = x_test.reshape(len(x_test),784)
x_test_final_flattened = x_test_final/255
x_test_final.shape
```

```python
plt.matshow(x_test[0])
```

```python
predict = model.predict(x_test_final_flattened)
```

```python
predict[0]
```

```python
output = np.argmax(predict[0])
print(output)
```
